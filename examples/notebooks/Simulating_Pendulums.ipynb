```python
import hephaestus as hp
from hephaestus.integrators.velocity_verlet import VelocityVerlet
from hephaestus.systems.single_pendulum import SinglePendulum
import matplotlib.pyplot as plt
import numpy as np
from scipy.fft import fft, ifft
from scipy.fft import fft, fftfreq
```

```python
N=10000
```

```python
pendulum = SinglePendulum(time_step=0.01, steps=N, mass=1.0, gravity=1.0, theta_start=5.0)
```

```python
integrator = VelocityVerlet(time_step=0.01, model=pendulum)
```

```python
pendulum.integrator = integrator
```

```python
pendulum.run_simulation()
```

```python
plt.plot(np.linspace(0, N, N), pendulum.pe, label='PE')
plt.plot(np.linspace(0, N, N), pendulum.ke, label='KE')
plt.legend()
plt.show()
```

```python
plt.plot(np.linspace(0, N, N), pendulum.ke, label='PE')
plt.plot(np.linspace(0, N, N), pendulum.velocity, label='vel')
plt.legend()
plt.show()
```

```python
plt.plot(np.linspace(0, N, N), pendulum.ke + pendulum.pe, label='tota')
plt.legend()
plt.show()
```
