```python
import hephaestus as hp
from hephaestus.integrators.velocity_verlet import VelocityVerlet
from hephaestus.systems.single_pendulum import SinglePendulum
import matplotlib.pyplot as plt
import numpy as np
```

```python
N=10000
```

```python
pendulum = SinglePendulum(time_step=0.001, steps=N, mass=1.0, gravity=10.0, theta_start=5.0)
```

```python
integrator = VelocityVerlet(time_step=0.001, model=pendulum)
```

```python
pendulum.integrator = integrator
```

```python
pendulum.run_simulation()
```

```python
plt.plot(np.linspace(0, N, N), pendulum.pe, label='PE')
plt.plot(np.linspace(0, N, N), pendulum.ke, label='KE')
plt.legend()
plt.show()
```

```python
plt.plot(np.linspace(0, N, N), pendulum.ke + pendulum.pe, label='total')
plt.legend()
plt.show()
```
